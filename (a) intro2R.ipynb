```r
name <- c("Jon","Bill","Maria")
age <- c(23,41,32)

df <- data.frame(name,age)

print(df)
```

```r
df <- data.frame(name = c("Jon","Bill","Maria"),
                age = c(23,41,32))

print(df)
```

#### Creating subsets of data frames with select and filter

```r
df
```

We can select the column and assign a variable for it.

```r
language_col <- select(df, language)
language_col
```

We can use "select" to extract multiple columns.

```r
three_col <- select(df, language,mother_tongue,most_at_home)
```

We can do the same as a range.

```r
three_col <- select(df, language : most_at_home)
```

We use "filter" to extract a single row.

```r
mandarin <- filter(df, language == "Mandarin")
mandarin
```

Using filter to extract rows with values above a threesold

```r
spoke_often_at_home <- filter(df, most_at_home > 462890)
spoke_often_at_home
```

### EXploring data with visualizations
#### Using ggplot to create a scatter plot

```r
ggplot(df, aes(x = most_at_home, y = mother_tongue))+ geom_point() + xlab("Language spoken at home") + 
ylab("Mother tongue")
```

aes  --> aesthetic function to specify axis

'+'  --> add a layer to the plot

geom_plot  --> scatter plot

#### Changing the units

```r
df <- mutate(df, mother_tongue = mother_tongue / 35151728)
df
```

```r
df <- mutate(df, most_at_home = most_at_home / 35151728)
df
```

```r
ggplot(df, aes(x = most_at_home, y = mother_tongue))+ geom_point() + xlab("Language spoken at home") + 
ylab("Mother tongue") + scale_x_log10(labels = scales::comma)+ scale_y_log10(labels = scales::comma)
```

### Coloring points by group

```r
ggplot(df, aes(x = most_at_home, y = mother_tongue, color = category))+ geom_point() + xlab("Language spoken at home") + 
ylab("Mother tongue") + scale_x_log10(labels = scales::comma)+ scale_y_log10(labels = scales::comma)
```
